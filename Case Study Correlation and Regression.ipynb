## 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
# columns present in the dataset
data.columns
```

```python
#dependent variable
y=data['Profit(in Thousands)']
```

```python
# independent variable
x=data.drop(['Date','Profit(in Thousands)'],axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
#split the Dataset into train and test
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
x_train.shape
```

```python
#perform linear regression
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
#make prediction using the model
predictions=model.predict(x_test)
```

```python
#output of the actual data
print( y_test)
```

```python
#output of the prediction
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('MSE is',mean_squared_error(y_test,predictions))
```

* MSE used to check difference between the observed and predicted values.
* smaller the MSE ,line of best fit.


```python
rsquare= r2_score(y_test,predictions)
print('R2 value:',r2_score(y_test,predictions))
```

 * r squared values used to meassure goodeness of fit.
 * Greater value of r square ,better the model. 
 * Here the r squared is 1,the model fits the data perfectly with a corresponding MSE.

```python
(sum(predictions)/sum(y_test))*100
```

* This model predicted with 99.9999% Accuracy.
